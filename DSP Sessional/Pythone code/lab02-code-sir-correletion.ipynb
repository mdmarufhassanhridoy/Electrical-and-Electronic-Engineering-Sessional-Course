```python
import numpy as np
import matplotlib.pyplot as plt

# Given signals
x = np.array([1, 2, 3, 4])
h = np.array([4, 3, 4, 3])

# Calculate auto-correlation of signal x
auto_corr_x = np.zeros(2 * len(x) - 1)
for lag in range(-len(x) + 1, len(x)):
    for i in range(len(x)):
        if i + lag >= 0 and i + lag < len(x):
            auto_corr_x[lag + len(x) - 1] += x[i] * x[i + lag]

# Calculate cross-correlation of signals x and h
cross_corr = np.zeros(2 * len(x) - 1)
for lag in range(-len(x) + 1, len(x)):
    for i in range(len(x)):
        if i + lag >= 0 and i + lag < len(h):
            cross_corr[lag + len(x) - 1] += x[i] * h[i + lag]

# Create time axis for correlation plots
time_axis = np.arange(-len(x) + 1, len(x))

# Plot auto-correlation of signal x
plt.figure(figsize=(10, 15))
plt.subplot(3, 1, 1)
plt.stem(time_axis, auto_corr_x)
plt.title('Auto-correlation of Signal x')
plt.xlabel('Time Lag')
plt.ylabel('Correlation')
plt.grid(True)

# Plot cross-correlation of signals x and h
plt.subplot(3, 1, 2)
plt.stem(time_axis, cross_corr)
plt.title('Cross-correlation of Signals x and h')
plt.xlabel('Time Lag')
plt.ylabel('Correlation')
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
