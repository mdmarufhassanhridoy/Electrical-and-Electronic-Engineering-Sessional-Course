```python
import numpy as np
import matplotlib.pyplot as plt

# Given signals
x = np.array([1, 2, 3, 4])
h = np.array([4, 3, 4, 3])

# Calculate the length of the convolution output
conv_len = len(x) + len(h) - 1

# Initialize the convolution output array
conv = np.zeros(conv_len)

# Perform the convolution using nested loops
for n in range(conv_len):
    for k in range(len(x)):
        if n - k >= 0 and n - k < len(h):
            conv[n] += x[k] * h[n - k]

# Create time axis for the convolution plot
time_axis = np.arange(0, conv_len)

# Plot the convolution result
plt.figure(figsize=(10, 5))
plt.stem(time_axis, conv)
plt.title('Convolution of Signals x and h')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
