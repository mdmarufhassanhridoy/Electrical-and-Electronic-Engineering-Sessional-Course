```python
import numpy as np
import matplotlib.pyplot as plt

# Create two signals
signal1 = np.random.rand(100)
signal2 = np.random.rand(100)

# Calculate autocorrelation of signal1
autocorr = np.correlate(signal1, signal1, mode='full')

# Calculate cross-correlation of signal1 and signal2
crosscorr = np.correlate(signal1, signal2, mode='full')

# Plot the signals and their autocorrelation/cross-correlation
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(signal1)
axs[0, 0].set_title('Signal 1')

axs[0, 1].plot(signal2)
axs[0, 1].set_title('Signal 2')

axs[1, 0].plot(autocorr)
axs[1, 0].set_title('Autocorrelation of Signal 1')

axs[1, 1].plot(crosscorr)
axs[1, 1].set_title('Cross-correlation of Signal 1 and Signal 2')

plt.show()
```
