```python
import numpy as np

x = np.array([-1, 2, 0, 1])  # Define the first sequence
h = np.array([3, 1, 0, -1])  # Define the second sequence

# Perform convolution using numpy's convolve function
y = np.convolve(x, h)

# Print the result
print("Convolution result:", y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-1, 2, 0, 1])  # Define the first sequence
h = np.array([3, 1, 0, -1])  # Define the second sequence

# Perform convolution using numpy's convolve function
y = np.convolve(x, h)

# Plotting the signals
plt.figure(figsize=(10, 6))

# Plot the first sequence
plt.subplot(3, 1, 1)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('First Sequence')

# Plot the second sequence
plt.subplot(3, 1, 2)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Second Sequence')

# Plot the convolution result
plt.subplot(3, 1, 3)
plt.stem(range(len(y)), y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Convolution Result')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-1, 2, 0, 1])  # Define the first sequence
h = np.array([3, 1, 0, -1])  # Define the second sequence

# Perform convolution using loops
y = np.zeros(len(x) + len(h) - 1)
for i in range(len(x)):
    for j in range(len(h)):
        y[i + j] += x[i] * h[j]

# Plotting the signals
plt.figure(figsize=(10, 6))

# Plot the first sequence
plt.subplot(3, 1, 1)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('First Sequence')

# Plot the second sequence
plt.subplot(3, 1, 2)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Second Sequence')

# Plot the convolution result
plt.subplot(3, 1, 3)
plt.stem(range(len(y)), y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Convolution Result')

plt.tight_layout()
plt.show()
```
