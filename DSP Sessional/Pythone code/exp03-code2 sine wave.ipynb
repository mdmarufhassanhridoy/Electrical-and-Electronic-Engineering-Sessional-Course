```python
import numpy as np
import matplotlib.pyplot as plt

# Given signals
t = np.linspace(0, 1, 1000)  # Time array
s1 = np.sin(2 * np.pi * 5 * t)  # Sine signal 1 (5 Hz)
s2 = np.sin(2 * np.pi * 10 * t)  # Sine signal 2 (10 Hz)

# Calculate s3 = s1 + s2
s3 = s1 + s2

# Plot the FFT of s1, s2, and s3
fft_s1 = np.fft.fft(s1)
fft_s2 = np.fft.fft(s2)
fft_s3 = np.fft.fft(s3)

freq = np.fft.fftfreq(len(t), 1 / len(t))

plt.figure(figsize=(10, 5))

plt.subplot(3, 1, 1)
plt.plot(freq, np.abs(fft_s1))
plt.title('FFT of s1')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(freq, np.abs(fft_s2))
plt.title('FFT of s2')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(freq, np.abs(fft_s3))
plt.title('FFT of s3')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# Add random noise to s3
noise = np.random.normal(0, 0.5, len(t))
s4 = s3 + noise

# Remove random noise from s4 using FFT analysis
fft_s4 = np.fft.fft(s4)
clean_fft_s4 = np.where(np.abs(fft_s4) > np.max(np.abs(fft_s4)) * 0.5, fft_s4, 0)
clean_s4 = np.fft.ifft(clean_fft_s4).real

# Denoise the signal s4 using a Butterworth filter
order = 4  # Butterworth filter order
cutoff_freq = 15  # Cutoff frequency for the filter

# Design the Butterworth filter
b, a = signal.butter(order, cutoff_freq, fs=2*np.pi)

# Apply the filter to s4
denoised_s4 = signal.filtfilt(b, a, s4)

# Plot the original s4, the FFT-based denoised s4, and the Butterworth-filtered s4
plt.figure(figsize=(10, 5))

plt.subplot(3, 1, 1)
plt.plot(t, s4)
plt.title('Original s4')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, clean_s4)
plt.title('FFT-based Denoised s4')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, denoised_s4)
plt.title('Butterworth-filtered s4')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
