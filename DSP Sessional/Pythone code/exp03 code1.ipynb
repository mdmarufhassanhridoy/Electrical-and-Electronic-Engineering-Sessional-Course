```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
from scipy.signal import butter, filtfilt

# Given signals
s1 = np.array([1, 2, 3, 4])
s2 = np.array([4, 3, 2, 1])

# Calculate s3 = s1 + s2
s3 = s1 + s2

# Plot the FFT of s1, s2, and s3
freq = np.fft.fftfreq(len(s1))
fft_s1 = np.fft.fft(s1)
fft_s2 = np.fft.fft(s2)
fft_s3 = np.fft.fft(s3)

plt.figure(figsize=(10, 5))

plt.subplot(3, 1, 1)
plt.plot(freq, np.abs(fft_s1))
plt.title('FFT of s1')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(freq, np.abs(fft_s2))
plt.title('FFT of s2')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(freq, np.abs(fft_s3))
plt.title('FFT of s3')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# Add random noise to s3
noise = np.random.normal(0, 0.5, len(s3))
s4 = s3 + noise

# Remove random noise from s4 using FFT analysis
fft_s4 = np.fft.fft(s4)
clean_fft_s4 = np.where(np.abs(fft_s4) > np.max(np.abs(fft_s4)) * 0.5, fft_s4, 0)
clean_s4 = np.fft.ifft(clean_fft_s4).real

# Denoise the signal s4 using a Butterworth filter
order = 4  # Butterworth filter order
cutoff_freq = 0.2  # Cutoff frequency for the filter

# Design the Butterworth filter
b, a = butter(order, cutoff_freq, btype='low', analog=False)

# Apply the filter to s4
denoised_s4 = filtfilt(b, a, s4)

# Plot the original s4, the FFT-based denoised s4, and the Butterworth-filtered s4
plt.figure(figsize=(10, 5))

plt.subplot(3, 1, 1)
plt.plot(s4)
plt.title('Original s4')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(clean_s4)
plt.title('FFT-based Denoised s4')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(denoised_s4)
plt.title('Butterworth-filtered s4')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
